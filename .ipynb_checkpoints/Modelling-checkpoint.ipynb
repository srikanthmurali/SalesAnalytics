```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load your dataset (replace 'your_dataset.csv' with your actual file)
data = pd.read_excel('ELGI.axlsx')
data
```

```python
# Define the features (X) and target variable (y)
X = data[['monthnum', 'month','year','categories', 'region', 'units', 'cost_per_unit', 'sales_cost', 'staff_qty', 'per_staff', 'staff_cost']]
y = data['Customer_type']  # Assuming 'Customer_Type' represents the customer types (1, 2, 3, 4)
```

```python
# Encode categorical features (if needed)
X_encoded = pd.get_dummies(X, columns=['categories', 'region', 'month'])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Create a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions for the dataset in 2023
# Prepare a similar dataset for 2023 with columns 'monthnum', 'year', 'units', 'cost_per_unit', 'sales_cost', 'staff_qty', 'per_staff', 'staff_cost'
# Use clf.predict() to make predictions on this dataset.


```

```python
# Calculate accuracy
accuracy = accuracy_score(y_train, predictions_2023)
print(f'Accuracy: {accuracy}')

# Generate a classification report for detailed metrics
report = classification_report(y_test, predictions_2023)
print(f'Classification Report:\n{report}')
```
